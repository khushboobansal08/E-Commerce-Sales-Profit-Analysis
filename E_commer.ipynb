```python
import pandas as pd

import plotly.express as px 
import plotly.graph_objects as gp
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("Sample - Superstore.csv",encoding = "latin-1")
```

#converting data type 

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
```

```python
data['Ship Date']=pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data['Postal Code']= data['Postal Code'].astype(int)
```

```python
data['Postal Code'].dtype
```

```python
data['Order Month']= data['Order Date'].dt.month
data['Order Year']= data['Order Date'].dt.year
data['Order day of week']= data['Order Date'].dt.dayofweek
```

```python
Sales_by_Month = data.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
Sales_by_Month 
```

```python
fig = px.line(Sales_by_Month ,
              x = 'Order Month',
              y = 'Sales',
              title = 'Sales By Month'
)
fig.show()
```

```python
Sales_by_Category = data.groupby('Category')['Sales'].sum().reset_index()
fig_1 = px.pie(
    Sales_by_Category,
    names="Category",
    values="Sales",
    hole = 0.5 ,
    title="Sales by Category"
)

fig_1.show()
```

```python
Sales_by_Category
```

```python
Sales_by_SubCategories = data.groupby('Sub-Category')['Sales'].sum().reset_index()
Sales_by_SubCategories
```

```python
fig_2 = px.bar (
    Sales_by_SubCategories,
    x = 'Sub-Category',
    y = 'Sales',
    title='Analysis'
)
```

```python
fig_2.show()
```

```python
Monthly_profit = data.groupby('Order Month')['Profit'].sum().reset_index()
Monthly_profit
```

```python
fig_3 = px.line(
    Monthly_profit,
    x='Order Month',
    y = 'Profit'
)
fig_3.show()
```

```python
Profit_by_Category= data.groupby('Category')['Profit'].sum().reset_index()
fig_4= px.pie(
    Profit_by_Category,
    names = 'Category',
    values = 'Profit',
    title='Profit by Category',
    color_discrete_sequence = px.colors.qualitative.Pastel
)
fig_4.update_traces(textposition = 'inside',textinfo ='percent+label')
fig_4.show()
```

```python
Profit_by_SubCategory= data.groupby('Sub-Category')['Profit'].sum().reset_index()
fig_5= px.bar(
    Profit_by_SubCategory,
    x = 'Sub-Category',
    y = 'Profit',
    title='Profit by Sub-Category'
)
fig_5.show()
```

```python
Sales_and_Profit_by_segment = data.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
fig_6 = gp.Figure()
color_palette= colors.qualitative.Pastel
fig_6.add_traces(gp.Bar(x= Sales_and_Profit_by_segment['Segment'],
                       y = Sales_and_Profit_by_segment['Sales'],
                       name= 'Sales',
                       marker_color= color_palette[0]))

fig_6.add_traces(gp.Bar(x= Sales_and_Profit_by_segment['Segment'],
                       y=Sales_and_Profit_by_segment['Profit'],
                       name= 'Profit',
                      marker_color= color_palette[1]))
fig_6.update_layout(title= 'Sales and Profit with customer segment ',
                    xaxis_title = 'Customer Segment',yaxis_title = 'Amount' )

fig_6.show()
```

```python
Sales_and_Profit_by_segment['sales_to_profit_ratio'] = Sales_and_Profit_by_segment['Sales']/Sales_and_Profit_by_segment['Profit']
print(Sales_and_Profit_by_segment[['Segment','sales_to_profit_ratio']])
```
